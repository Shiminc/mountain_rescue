```python
# use data from 2012 as most months will have incident
# add 0 for the following
# 2017 - 12
# 2020 - 4
# 2021 - 3
#
data_pd = data_pd[data_pd['year']>2011]
incident_count = pd.DataFrame(data_pd.groupby(by =['year_month','year','month'])['Incident'].count()).reset_index()
incident_count
df_zero = pd.DataFrame({"year_month":['2017-12-01', '2020-04-01', '2021-03-01'],
                    "year":[2017, 2020, 2021],
                    "month":[12, 4, 3],
                    "Incident":[0,0,0]})
```

```python
data_pd.to_csv('jun.csv')
```

```python
data_graph = pd.concat([incident_count, df_zero]).reset_index()
```

```python
data_graph['year_month']= pd.to_datetime(data_graph['year_month'], format = "%Y-%m-%d")
data_graph.sort_values(by=['year','month'],inplace=True)
```

```python
year_ave = pd.DataFrame(data_graph.groupby(by='year')['Incident'].agg('mean')).reset_index()
month_ave = pd.DataFrame(data_graph.groupby(by='month')['Incident'].agg('mean')).reset_index()
year_ave.rename(columns={"Incident":"Year_Average"}, inplace=True)
month_ave.rename(columns={"Incident":"Month_Average"}, inplace=True)
```

```python
data_graph = data_graph.merge(year_ave, left_on='year', right_on='year')
data_graph = data_graph.merge(month_ave, left_on='month', right_on='month')
```

```python
data_hrs['start_time_obj'] = pd.to_datetime(data_hrs['start_time'], format = "%H:%M")
data_hrs['end_time_obj'] = pd.to_datetime(data_hrs['end_time'], format = "%H:%M")
```

```python
data_hrs['next_day'] = data_hrs['end_time_obj'] < data_hrs['start_time_obj']   
data_hrs['end_time_calculate'] = data_hrs['end_time_obj']
data_hrs.loc[data_hrs['next_day'] == True,'end_time_calculate'] =  data_hrs['end_time_obj'] +  pd.Timedelta(days=1)
```

```python
data_hrs['time_used'] = data_hrs['end_time_calculate'] - data_hrs['start_time_obj']
```

```python
base = alt.Chart(data[['start_time', 'end_time', 'start_time_obj','end_time_calculate', 'staff', 'title', 'date', 'Incident_Cause']])
#axis=alt.Axis(title=None,labels=False)

chart_start = base.mark_point().encode(
    x = alt.X('yearmonthdatehoursminutes(start_time_obj):T', axis = alt.Axis(labels=False), scale=alt.Scale(domain=['1900-01-01T00:00:00', '1900-01-03T00:00:00'])
    ).title(None),
    y = alt.Y('date'),
    tooltip = [
            alt.Tooltip(field="start_time"),
            alt.Tooltip(field="end_time")    
        ],
)

chart_end = chart_start.encode(
    x = alt.X('yearmonthdatehoursminutes(end_time_calculate):T', axis = alt.Axis(labels=False), scale=alt.Scale(domain=['1900-01-01T00:00:00', '1900-01-03T00:00:00'])
).title(None),
    
)

chart_line = base.mark_rule().encode(
    x = alt.X('start_time_obj', axis = alt.Axis(labels=False)).title(None),
    y = alt.Y('date'),
    x2 = 'end_time_calculate',
    strokeWidth='staff:Q',
    color = 'Incident_Cause:N'
)

# mark the next day
rule = base.mark_rule(strokeDash=[2, 2]).encode(
    x=alt.datum(alt.DateTime(year=1900, month = 1, date = 2, hours = 0)),
    color=alt.value("red")
)

chart_start + chart_end + chart_line + rule
```
