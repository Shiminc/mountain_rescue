```python
url ='https://www.wmrt.org.uk/incidents/upper-eskdale-sat-18th-jan-2025/'
page_content = requests.get(url)
soup = BeautifulSoup(page_content.text, 'html.parser')
if soup.find('div', dir = 'auto'):
    print('t')
```

```python
extract_data = []
title = soup.find('h1').get_text()

table = soup.find('dl','dl-horizontal')
headings = table.find_all('dt')
details = table.find_all('dd')

rescue_info = soup.find(id="rescueinfo")
extract_data.append({'title': title})

```

```python
matches = re.search(r"([0-9]{1,2})(?:st|nd|rd|th) ([a-zA-Z]{3} [0-9]{4})", title)
matches.group(1) + ' ' +  matches.group(2)
```

```python
table_dict = {'title':title}
for i in range(0,len(headings)):
    table_dict[headings[i].string] = details[i].get_text().replace('\xa0', ' ')
    
```

```python
table_dict
```

```python
if 'Type' in table_dict.keys():
    incident = table_dict['Type'].split(' — ')
    table_dict['Incident_Type'] = incident[0]
    table_dict['Incident_Cause'] = incident[1]

if 'Weather' in table_dict.keys():
    table_dict['Weather'] = table_dict['Weather'].split(', ')

if 'Tagged' in table_dict.keys():
    table_dict['Tagged'] = table_dict['Tagged'].split(', ')

```

```python
matches = re.search(r"started at ([0-9]{2}:[0-9]{2})", information) 
if matches:
    start_time = matches.group(1)
    print(start_time)
matches = re.search(r"ended at ([0-9]{2}:[0-9]{2})", information) 
if matches:
    end_time = matches.group(1)
    print(end_time)
matches = re.search(r"(\d+\.?\d*) hrs", information)
if matches:
    hrs = matches.group(1)
    print(hrs)
matches = re.search(r"(\d+) Wasdale", information) 
if matches:
    staff = matches.group(1)
    print(staff)
matches = re.search(r"(NY[ \d]+)", information) 
if matches:
    location = matches.group(1)
    print(location)
matches = re.search(r"Total rescuer hours: (\d+\.?\d*)", information)
if matches:
    total_hrs = matches.group(1)
    print(total_hrs)
```
